```python
url = 'http://ergast.com/api/f1/2008/5/driverStandings.json'
```

```python
#response object
response = requests.get(url)
```

```python
#check and make sure we got a successful response from the API

if response.status_code == 200:
    data = response.json()
    print("Successfully retrieved data:")
    print(data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
#view the response as a JSON
if response.status_code == 200:
    data = response.json()
    print("Successfully retrieved data:")
    print(data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
# We only want the Driver Standings
```

```python
if response.status_code == 200:
    data = response.json()
    
    standings = data.get('MRData', {}).get('StandingsTable', {}).get('StandingsLists', [])
    
    if standings:
        print("Driver Standings:")
        for standing in standings:
            for driver in standing.get('DriverStandings', []):
                driver_info = driver.get('Driver', {})
                print(f"Position: {driver.get('position')}")
                print(f"Driver: {driver_info.get('givenName')} {driver_info.get('familyName')}")
                print(f"Points: {driver.get('points')}")
                print(f"Nationality: {driver_info.get('nationality')}")
                print("-" * 30)
    else:
        print("No driver standings data available.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
    print("Response content:", response.text)
```

```python
import requests
import json

def get_pokemon_data(pokemon_id):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_id}/'
    response = requests.get(url)
    data = response.json()
    return data

def categorize_pokemon(num_pokemon):
    pokemon_dict = {}

    for i in range(1, num_pokemon + 1):
        data = get_pokemon_data(i)
        name = data['name']
        abilities = [ability['ability']['name'] for ability in data['abilities']]
        weight = data['weight'] / 10  # Convert hectograms to kilograms
        types = [t['type']['name'] for t in data['types']]
        
        for pokemon_type in types:
            if pokemon_type not in pokemon_dict:
                pokemon_dict[pokemon_type] = {}
            pokemon_dict[pokemon_type][name] = {
                'abilities': abilities,
                'weight': weight
            }

    return pokemon_dict
pokemon_data = categorize_pokemon(20)
print(json.dumps(pokemon_data, indent=2))
```
